```python
from statsmodels.stats.diagnostic import het_arch

def rolling_arch_test(series: pd.Series, window: int = 60, nlags: int = 5):
    stat_list, pval_list, r2_list = [], [], []

    for i in range(window, len(series)):
        window_data = series.iloc[i - window:i]
        try:
            lm_stat, lm_pval, _, _ = het_arch(window_data, nlags=nlags)
            r2 = np.corrcoef(window_data**2, np.roll(window_data**2, 1))[0, 1]**2
        except Exception as e:
            print(e)
            lm_stat, lm_pval, r2 = np.nan, np.nan, np.nan

        stat_list.append(lm_stat)
        pval_list.append(lm_pval)
        r2_list.append(r2)

    return pd.Series(stat_list, index=series.index[window:]), \
           pd.Series(pval_list, index=series.index[window:]), \
           pd.Series(r2_list, index=series.index[window:])
```

```python
import numpy as np
import pandas as pd
from numba import njit
from scipy.stats import chi2, f

@njit(cache=True, fastmath=True)
def _fast_het_arch_numba(y: np.ndarray, nlags: int, ddof: int = 0) -> tuple:
    nobs = y.shape[0] - nlags
    if nobs <= nlags + 1:
        return np.nan, np.nan, np.nan, np.nan

    y_target = y[nlags:]
    y_lagged = np.empty((nobs, nlags))
    for i in range(nlags):
        y_lagged[:, i] = y[nlags - i - 1: -i - 1]

    X = np.ones((nobs, nlags + 1))
    X[:, 1:] = y_lagged

    XtX = X.T @ X
    Xty = X.T @ y_target
    beta = np.linalg.solve(XtX, Xty)
    y_hat = X @ beta
    resid = y_target - y_hat

    rss = np.dot(resid, resid)
    tss = np.dot(y_target - y_target.mean(), y_target - y_target.mean())
    r_squared = 1 - rss / tss

    lm_stat = (nobs - ddof) * r_squared
    mse_model = np.dot(y_hat - y_target.mean(), y_hat - y_target.mean()) / nlags
    mse_resid = rss / (nobs - nlags - 1)
    f_stat = mse_model / mse_resid

    return lm_stat, r_squared, f_stat, nobs

def fast_het_arch_numba(resid: np.ndarray, nlags: int = 5, ddof: int = 0) -> tuple:
    y = resid ** 2
    lm_stat, r_squared, f_stat, nobs = _fast_het_arch_numba(y, nlags, ddof)

    if np.isnan(lm_stat):
        return np.nan, np.nan, np.nan, np.nan

    lm_pval = chi2.sf(lm_stat, df=nlags)
    f_pval = f.sf(f_stat, dfn=nlags, dfd=nobs - nlags - 1)
    return lm_stat, lm_pval, f_stat, f_pval

def f_rolling_arch_test(df: pd.DataFrame, col: str, window: int = 60, nlags: int = 5, ddof: int = 0) -> pd.DataFrame:
    """
    Apply ARCH test in a rolling window fashion using a fast Numba implementation.

    Parameters
    ----------
    df : pd.DataFrame
        Input DataFrame containing returns or residuals.
    col : str
        Column name to apply the ARCH test on.
    window : int
        Rolling window size.
    nlags : int
        Number of lags in the ARCH test.
    ddof : int
        Degrees of freedom used in the LM statistic.

    Returns
    -------
    pd.DataFrame
        DataFrame with columns ['arch_lm', 'arch_lm_pval', 'arch_f', 'arch_f_pval', 'arch_r2'].
    """
    lm_stats = []
    lm_pvals = []
    f_stats = []
    f_pvals = []
    r2_list = []
    index = df.index[window:]

    for i in range(window, len(df)):
        window_data = df[col].iloc[i - window:i].values
        y = window_data ** 2
        lm_stat, r_squared, f_stat, nobs = _fast_het_arch_numba(y, nlags, ddof)

        if np.isnan(lm_stat):
            lm_stats.append(np.nan)
            lm_pvals.append(np.nan)
            f_stats.append(np.nan)
            f_pvals.append(np.nan)
            r2_list.append(np.nan)
            continue

        lm_pval = chi2.sf(lm_stat, df=nlags)
        f_pval = f.sf(f_stat, dfn=nlags, dfd=nobs - nlags - 1)

        lm_stats.append(lm_stat)
        lm_pvals.append(lm_pval)
        f_stats.append(f_stat)
        f_pvals.append(f_pval)
        r2_list.append(r_squared)

    return pd.DataFrame({
        "arch_lm": lm_stats,
        "arch_lm_pval": lm_pvals,
        "arch_f": f_stats,
        "arch_f_pval": f_pvals,
        "arch_r2": r2_list
    }, index=index)
```

```python
import numpy as np
import pandas as pd
from numba import njit
from scipy.stats import chi2

@njit(cache=True, fastmath=True)
def _arch_lm_only(y: np.ndarray, nlags: int, ddof: int = 0) -> float:
    nobs = y.shape[0] - nlags
    if nobs <= nlags + 1:
        return np.nan

    y_target = y[nlags:]
    y_lagged = np.empty((nobs, nlags))
    for i in range(nlags):
        y_lagged[:, i] = y[nlags - i - 1: -i - 1]

    X = np.ones((nobs, nlags + 1))
    X[:, 1:] = y_lagged

    XtX = X.T @ X
    Xty = X.T @ y_target
    beta = np.linalg.solve(XtX, Xty)
    y_hat = X @ beta
    resid = y_target - y_hat

    rss = np.dot(resid, resid)
    tss = np.dot(y_target - y_target.mean(), y_target - y_target.mean())
    r_squared = 1 - rss / tss

    return (nobs - ddof) * r_squared


def arch_test(df: pd.DataFrame, col: str, window_size: int = 60, lags: int = 5, ddof: int = 0) -> tuple[pd.Series, pd.Series]:
    """
    Compute the ARCH test (Engle) over rolling windows to detect conditional heteroskedasticity.

    This function applies the ARCH Lagrange Multiplier test in a rolling fashion
    to a given time series. It returns both the LM statistic and the associated p-value.
    The ARCH test measures whether volatility is autocorrelated (i.e., clustering),
    which is common in financial time series.

    Parameters
    ----------
    df : pd.DataFrame
        Input DataFrame containing the time series data.
    col : str
        Name of the column to test (typically returns or residuals).
    window_size : int, optional (default=60)
        Size of the rolling window used to estimate ARCH effects.
    lags : int, optional (default=5)
        Number of lags to include in the ARCH regression (squared residuals).
    ddof : int, optional (default=0)
        Degrees of freedom adjustment (useful when residuals come from a fitted model).

    Returns
    -------
    arch_stat : pd.Series
        Rolling series of the LM statistics from the ARCH test.
    arch_pval : pd.Series
        Rolling series of the associated p-values (under Chi2 distribution).

    Raises
    ------
    ValueError
        If inputs are invalid: missing column, non-numeric data, or incorrect parameters.
    """

    # --- Input validation ---
    if col not in df.columns:
        raise ValueError(f"Column '{col}' not found in the DataFrame.")

    if not np.issubdtype(df[col].dtype, np.number):
        raise ValueError(f"Column '{col}' must contain numeric values.")

    if not isinstance(window_size, int) or window_size <= 1:
        raise ValueError(f"'window_size' must be a positive integer > 1. Got {window_size}.")

    if lags is not None and (not isinstance(lags, int) or lags < 1):
        raise ValueError(f"'lags' must be an integer >= 1. Got {lags}.")

    if not isinstance(ddof, int) or ddof < 0:
        raise ValueError(f"'ddof' must be a non-negative integer. Got {ddof}.")

    # --- Determine nlags ---
    if lags is None:
        nlags = int(np.floor(12 * (window_size / 100) ** 0.25))  # Schwert's rule
    else:
        nlags = lags

    if window_size <= nlags + 1:
        raise ValueError(f"'window_size' must be greater than 'lags + 1' for regression to be valid.")

    # --- Rolling ARCH computation ---
    lm_stats = []
    index = df.index[window_size:]

    for i in range(window_size, len(df)):
        window_data = df[col].iloc[i - window_size:i].values
        y = window_data ** 2
        lm_stat = _arch_lm_only(y, nlags, ddof)
        lm_stats.append(lm_stat)

    lm_stats = np.array(lm_stats)
    lm_pvals = chi2.sf(lm_stats, df=nlags)

    return (
        pd.Series(lm_stats, index=index, name="arch_stat"),
        pd.Series(lm_pvals, index=index, name="arch_pval")
    )

```

```python
def rolling_skewness(df: pd.DataFrame, col: str, window_size: int = 60) -> pd.Series:
    """
    Compute the skewness (third standardized moment) over a rolling window.

    Parameters
    ----------
    df : pd.DataFrame
        Input DataFrame containing the time series data.
    col : str
        Name of the column to compute skewness on.
    window_size : int, optional (default=60)
        Number of periods for the rolling window.

    Returns
    -------
    pd.Series
        Rolling skewness of the specified column.

    Examples
    --------
    >>> df["skew"] = rolling_skewness(df, col="returns", window_size=50)
    """
    if col not in df.columns:
        raise ValueError(f"Column '{col}' not found in DataFrame.")
    if not np.issubdtype(df[col].dtype, np.number):
        raise ValueError(f"Column '{col}' must be numeric.")
    if not isinstance(window_size, int) or window_size <= 0:
        raise ValueError(f"'window_size' must be a positive integer. Got {window_size}.")

    return df[col].rolling(window=window_size).skew().rename("skewness")



def rolling_kurtosis(df: pd.DataFrame, col: str, window_size: int = 60) -> pd.Series:
    """
    Compute the kurtosis (fourth standardized moment) over a rolling window.

    Parameters
    ----------
    df : pd.DataFrame
        Input DataFrame containing the time series data.
    col : str
        Name of the column to compute kurtosis on.
    window_size : int, optional (default=60)
        Number of periods for the rolling window.

    Returns
    -------
    pd.Series
        Rolling kurtosis of the specified column.

    Examples
    --------
    >>> df["kurtosis"] = rolling_kurtosis(df, col="returns", window_size=50)
    """
    if col not in df.columns:
        raise ValueError(f"Column '{col}' not found in DataFrame.")
    if not np.issubdtype(df[col].dtype, np.number):
        raise ValueError(f"Column '{col}' must be numeric.")
    if not isinstance(window_size, int) or window_size <= 0:
        raise ValueError(f"'window_size' must be a positive integer. Got {window_size}.")

    return df[col].rolling(window=window_size).kurt().rename("kurtosis")

```

```python
from scipy.stats import norm, chi2

@njit(cache=True, fastmath=True)
def _runs_test_stat(signs: np.ndarray) -> float:
    n = signs.size
    if n == 0:
        return np.nan

    n_pos = 0
    n_neg = 0
    runs = 1  # at least one run if not empty

    for i in range(n):
        if signs[i] == 1:
            n_pos += 1
        else:
            n_neg += 1

        if i > 0 and signs[i] != signs[i - 1]:
            runs += 1

    if n_pos == 0 or n_neg == 0:
        return np.nan  # no variation => undefined test

    expected_runs = (2 * n_pos * n_neg) / (n_pos + n_neg) + 1
    variance_runs = (2 * n_pos * n_neg * (2 * n_pos * n_neg - n_pos - n_neg)) / \
                    (((n_pos + n_neg) ** 2) * (n_pos + n_neg - 1))

    z = (runs - expected_runs) / np.sqrt(variance_runs)
    return z


def runs_test(df: pd.DataFrame, col: str, window_size: int) -> tuple[pd.Series, pd.Series]:
    """
    Apply the one-sample runs test in a rolling window fashion.

    This non-parametric test evaluates the randomness of a binary sequence derived
    from the input series. A significant result suggests serial dependence or clustering.

    Parameters
    ----------
    df : pd.DataFrame
        DataFrame containing the time series data.
    col : str
        Name of the column on which to apply the runs test.
    window_size : int
        Size of the rolling window.

    Returns
    -------
    tuple[pd.Series, pd.Series]
        A tuple with:
        - Rolling Z-statistic from the runs test.
        - Corresponding p-values assuming normality (two-tailed).
    """
    if col not in df.columns:
        raise ValueError(f"Column '{col}' not found in the DataFrame.")

    if not np.issubdtype(df[col].dtype, np.number):
        raise ValueError(f"Column '{col}' must be numeric.")

    if not isinstance(window_size, int) or window_size <= 2:
        raise ValueError(f"'window_size' must be an integer > 2. Got {window_size}.")

    def runs_wrapper(x):
        cutoff = np.mean(x)  # same as statsmodels default
        binary_seq = (x >= cutoff).astype(np.int32)
        print(binary_seq)
        return _runs_test_stat(binary_seq)

    series = df[col]
    z_stats = series.rolling(window=window_size).apply(runs_wrapper, raw=True)
    p_vals = 2 * (1 - norm.cdf(np.abs(z_stats)))  # two-tailed p-value

    return z_stats.rename("runs_stat"), pd.Series(p_vals, index=z_stats.index, name="runs_pval")
```
