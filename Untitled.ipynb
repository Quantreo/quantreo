```python
import pandas as pd
import numpy as np
import antropy as ant


def rolling_sample_entropy(df: pd.DataFrame, col: str = "close", window_size: int = 100, order: int = 2) -> pd.Series:
    """
    Compute the rolling Sample Entropy of a time series column.

    This function applies the sample entropy measure over a rolling window,
    providing a dynamic view of the signal's irregularity over time.

    This module uses AntroPy (BSD-3-Clause) for entropy calculations.
    AntroPy © 2018–2025 Raphael Vallat — https://github.com/raphaelvallat/antropy
    """
    if window_size < 10:
        raise ValueError("Sample entropy requires window_size >= 10.")

    if order < 1:
        raise ValueError("Parameter 'order' must be >= 1.")

    return df[col].rolling(window=window_size).apply(
        lambda x: ant.sample_entropy(x, order=order) if len(x) == window_size else np.nan,
        raw=True
    )


def rolling_spectral_entropy(df: pd.DataFrame, col: str = "close", window_size: int = 100, sf: int = 1,
                             method: str = 'welch', normalize: bool = True, nperseg: int = None) -> pd.Series:
    """
    Compute the rolling Spectral Entropy of a time series column.

    Spectral entropy measures the flatness (complexity) of the power spectral density.

    This module uses AntroPy (BSD-3-Clause) for entropy calculations.
    AntroPy © 2018–2025 Raphael Vallat — https://github.com/raphaelvallat/antropy
    """
    if window_size < 16:
        raise ValueError("Spectral entropy requires window_size >= 16 for stable estimation.")

    if sf <= 0:
        raise ValueError("Sampling frequency 'sf' must be strictly positive.")

    if method not in ["welch", "fft"]:
        raise ValueError("Method must be 'welch' or 'fft'.")

    def compute_entropy(x):
        local_nperseg = nperseg if nperseg is not None else min(window_size, window_size // 2)
        return ant.spectral_entropy(x, sf=sf, method=method, normalize=normalize, nperseg=local_nperseg)

    return df[col].rolling(window=window_size).apply(
        lambda x: compute_entropy(x) if len(x) == window_size else np.nan,
        raw=True
    )


def rolling_permutation_entropy(df: pd.DataFrame, col: str = "close", window_size: int = 100, order: int = 3,
                                delay: int = 1, normalize: bool = True) -> pd.Series:
    """
    Compute the rolling Permutation Entropy of a time series column.

    Permutation entropy quantifies the complexity of the temporal order of values.

    This module uses AntroPy (BSD-3-Clause) for entropy calculations.
    AntroPy © 2018–2025 Raphael Vallat — https://github.com/raphaelvallat/antropy
    """
    if window_size < 10:
        raise ValueError("Permutation entropy requires window_size >= 10.")

    if order < 2:
        raise ValueError("Embedding 'order' must be >= 2.")

    if delay < 1:
        raise ValueError("Delay must be >= 1.")

    return df[col].rolling(window=window_size).apply(
        lambda x: ant.perm_entropy(x, order=order, delay=delay, normalize=normalize)
        if len(x) == window_size else np.nan,
        raw=True
    )


def rolling_detrended_fluctuation(df: pd.DataFrame, col: str = "close", window_size: int = 100,
                                  order: int = 1, nvals: np.ndarray = None, overlap: bool = False) -> pd.Series:
    """
    Compute the rolling Detrended Fluctuation Analysis (DFA) exponent of a time series column.

    DFA estimates long-term memory or fractal scaling by analyzing fluctuations across scales.

    This module uses AntroPy (BSD-3-Clause) for entropy calculations.
    AntroPy © 2018–2025 Raphael Vallat — https://github.com/raphaelvallat/antropy
    """
    if window_size < 100:
        raise ValueError("DFA requires window_size >= 100 for stable estimation.")

    if order < 1:
        raise ValueError("Polynomial 'order' must be >= 1.")

    return df[col].rolling(window=window_size).apply(
        lambda x: ant.detrended_fluctuation(x, order=order, nvals=nvals, overlap=overlap)
        if len(x) == window_size else np.nan,
        raw=True
    )


def rolling_petrosian_fd(df: pd.DataFrame, col: str = "close", window_size: int = 100) -> pd.Series:
    """
    Compute the rolling Petrosian Fractal Dimension of a time series column.

    Petrosian FD measures signal complexity using changes in directional derivatives.

    This module uses AntroPy (BSD-3-Clause) for entropy calculations.
    AntroPy © 2018–2025 Raphael Vallat — https://github.com/raphaelvallat/antropy
    """
    if window_size < 10:
        raise ValueError("Petrosian fractal dimension requires window_size >= 10.")

    return df[col].rolling(window=window_size).apply(
        lambda x: ant.petrosian_fd(x) if len(x) == window_size else np.nan,
        raw=True
    )
```
