```python
# Import a dataset to test the functions and create new ones easily
from quantreo.datasets import load_generated_ohlcv
df = load_generated_ohlcv()

df["returns"] = df["close"].pct_change(1)
# Show the data
df = df.dropna()
df
```

```python
from statsmodels.sandbox.stats.runs import runstest_1samp

def rolling_runs_z(series: np.ndarray) -> float:
    """
    Compute the Z-statistic of the runs test in a rolling window.

    Returns
    -------
    float
        Z-statistic from the runs test. Returns np.nan if an error occurs.
    """
    try:
        x = np.sign(series)
        z_stat, _ = runstest_1samp(x[~np.isnan(x)], correction=True)
        return z_stat
    except Exception as e:
        print(e)
        return np.nan

def rolling_runs_p(series: np.ndarray) -> float:
    """
    Compute the p-value of the runs test in a rolling window.

    Returns
    -------
    float
        p-value from the runs test. Returns np.nan if an error occurs.
    """
    try:
        x = np.sign(series)
        _, p_val = runstest_1samp(x[~np.isnan(x)], correction=True)
        return p_val
    except Exception as e:
        print(e)
        return np.nan
```

```python
%%time
z_ref = df["returns"].rolling(20).apply(rolling_runs_z, raw=True)
p_ref = df["returns"].rolling(20).apply(rolling_runs_p, raw=True)
```

```python
import numpy as np
from numba import njit
import pandas as pd
from scipy.stats import norm
import numpy as np
from numba import njit
from scipy.stats import norm

@njit(cache=True, fastmath=True)
def _runs_z_sm(seq: np.ndarray, correction: bool = True) -> float:
    """
    Runs‑test Z‑statistic (0/1) — copie exacte de statsmodels.runstest_1samp.
    """
    n = seq.size
    if n < 2:
        return np.nan

    # -------- même comptage des runs que statsmodels -----------
    runstart = np.nonzero(np.diff(np.concatenate((np.array([-np.inf]), seq, np.array([np.inf])))))[0]
    runs_len = np.diff(runstart)          # longueurs des runs
    runs_sign = seq[runstart[:-1]]        # signe de chaque run
    runs_pos = runs_len[runs_sign == 1]
    runs_neg = runs_len[runs_sign == 0]

    n_runs = runs_len.size
    n_pos = runs_pos.sum()                # somme des longueurs, pas simple count
    n_neg = runs_neg.sum()
    if n_pos == 0 or n_neg == 0:
        return np.nan

    npn = n_pos * n_neg
    r_mean = 2.0 * npn / n + 1.0
    r_var  = 2.0 * npn * (2.0 * npn - n) / (n**2 * (n - 1))
    if r_var <= 0.0:
        return np.nan

    z_num = n_runs - r_mean

    # continuity‑correction SAS
    if correction and n < 50:
        if z_num > 0.5:
            z_num -= 0.5
        elif z_num < 0.5:
            z_num += 0.5
        else:
            z_num = 0.0

    return z_num / np.sqrt(r_var)


def runs_test(df: pd.DataFrame, col: str, window_size: int,
              cutoff: str = "mean", correction: bool = True) -> tuple[pd.Series, pd.Series]:
    """
    Rolling runs‑test (identique à statsmodels).
    """
    # vérifications
    if col not in df.columns:
        raise ValueError(f"Column '{col}' not in DataFrame.")
    if not np.issubdtype(df[col].dtype, np.number):
        raise ValueError(f"Column '{col}' must be numeric.")
    if window_size < 3:
        raise ValueError("'window_size' must be ≥ 3.")

    def _wrapper(arr):
        valid = arr[np.isfinite(arr)]
        if valid.size < 3:
            return np.nan
        thr = valid.mean() if cutoff == "mean" else np.median(valid)
        seq = (valid >= thr).astype(np.int8)
        return _runs_z_sm(seq, correction)

    z = df[col].rolling(window_size).apply(_wrapper, raw=True)
    p = pd.Series(2 * (1 - norm.cdf(np.abs(z))), index=z.index)
    return z.rename("runs_stat"), p.rename("runs_pval")
```

```python
%%time
z_stats, p_vals = runs_test(df, col="returns", window_size=20)

```
